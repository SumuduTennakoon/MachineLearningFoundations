# Probability Distributions

## Coin Toss Simulation

```python
def coin_toss():
    choices = np.array([-1,1])
    return np.random.choice(choices)

def experiment(n):
    result = []
    for i in range(n):
        result.append(coin_toss())
    return np.array(result)

results = experiment(1000)
```

```python
import matplotlib.pyplot as plt

plt.hist(results, bins=2, ec="black")
plt.xlabel("score")
plt.ylabel("frequency")
plt.show() 
```

```python
def two_coin_experiment(n):
    result = []
    for i in range(n):
        result.append(coin_toss()+coin_toss())
    return np.array(result)

results = two_coin_experiment(1000)

plt.hist(results, bins=3, ec="black")
plt.xlabel("score")
plt.ylabel("frequency")
plt.show() 
```

```python
def three_coin_experiment(n):
    result = []
    for i in range(n):
        result.append(coin_toss()+coin_toss()+coin_toss())
    return np.array(result)

results = three_coin_experiment(1000)

plt.hist(results, bins=4, ec="black")
plt.xlabel("score")
plt.ylabel("frequency")
plt.show() 
```

# Descriptive Statistics with Numpy

## 1. Example: Analysing Exam Scores of 20 Students

```python
scores = np.array([75, 59, 63, 51, 53, 39, 85, 69, 73, 73, 34, 56, 34, 80, 78, 100, 64, 95, 74, 95])
print(F"scores = {scores}")
print(F"type(scores) = {type(scores)}")
```

```python
import matplotlib.pyplot as plt

plt.hist(scores, bins=5)
plt.xlabel("score")
plt.ylabel("frequency")
plt.show() 
```

### 1.1 Using Numpy Functions to get Descriptive Statistics

```python
print(F"min(scores) = {scores.min()}")
print(F"mean(scores) = {scores.mean()}")
print(F"std(scores) = {scores.std()}")
print(F"max(scores) = {scores.max()}")
print(F"range(scores) = {scores.max()-scores.min()}")
```

### 1.2 Find the index of maximum and minimum values

```python
print(F"index of min value = {scores.argmin()}") #array index starts from 0.
print(F"index of max value = {scores.argmax()}")
```

### 1.3 Calculating Z-Score
$ z = \frac{x_i- \mu}{\sigma} $

```python
z_scores = (scores - scores.mean())/scores.std()
print(F"z_scores = {z_scores}")
```

```python
# Round valeus upto 3rd decimal place

z_scores = np.round(z_scores, 3)
print(F"z_scores = {z_scores}")
```

### 1.4 Sorting Data using Numpy Array
https://numpy.org/doc/stable/reference/generated/numpy.sort.html

```python
print(F"scores un-sorted = {scores}")
print(F"scores sorted ascending = {np.sort(scores)}")
print(F"indices sorted ascending = {np.argsort(scores)}")
```

```python
print(F"cumulative sum ordered = {np.cumsum(np.sort(scores))}")
```

### 1.5 Getting Ranks

```python
# Create a zero array
ranks = np.zeros(len(scores), dtype='int')
ranks
```

```python
sorted_index = np.argsort(scores) #sorted asceding
sorted_index = np.flip(sorted_index) # get decending order (heighest to lowest score)
sorted_index
```

```python
# Generate rank labels
rank_labels = np.arange(1, len(scores)+1, 1)
rank_labels
```

```python
# Assign rank labales
ranks[sorted_index] = rank_labels
ranks
```

### 1.6 Creating Pandas Data Frame From Related Numpy Arrays

```python
score_table = pd.DataFrame(data={"score":scores, "z_score": z_scores, "rank":ranks})
score_table
```

### 1.7 Create ID Column

```python
student_ids = list(score_table.index+1)
student_ids
```

```python
score_table['student_id'] = student_ids
score_table[['student_id', 'score', 'z_score', 'rank']]
```

# Exercise 2

* Extend Coins toss similation to an experiemnt rolling two dice.

```python
def role_dice():
    choices = np.array([1,2,3,4,5,6])
    return np.random.choice(choices)

def experiment(n):
    result = []
    for i in range(n):
        result.append(role_dice())
    return np.array(result)

results = experiment(1000)

plt.hist(results, bins=6, ec="black")
plt.xlabel("score")
plt.ylabel("frequency")
plt.show() 

def two_dice_experiment(n):
    result = []
    for i in range(n):
        result.append(role_dice()+role_dice())
    return np.array(result)

results = two_dice_experiment(1000)

plt.hist(results, bins=11, ec="black")
plt.xlabel("score")
plt.ylabel("frequency")
plt.show() 
```
